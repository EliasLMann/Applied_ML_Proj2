```python
# added column with winner of each game
# H - home team, A - away team, D - draw
winner = []
scores = epl['Score']
for i in scores:
    score = i
    home_score = int(score[0])
    away_score = int(score.partition('-')[2])
    if home_score > away_score: winner.append(0)
    elif away_score > home_score: winner.append(2)
    else: winner.append(1)
```

```python
epl_reduced.insert(2, 'Winner',winner)
```

### Part 3 - Merge the data

Now that you have a better feel for each of your two (or three, for the 7394 students) data sets, it is time to merge them. Describe your strategy for merging the data sets and then actually perform the merge.  

Develop a strategy for verifying that the data is properly merged (hoping and finger-crossing are not valid strategies). 

```python
money['Home Team'] = money['Team']
money['Away Team'] = money['Team']
merged = pd.merge(epl_reduced, money[['Home Team', 'TM Value', 'year']], on=['Home Team', 'year'], how='left').rename(columns={'TM Value':'tm_h'})
merged = pd.merge(merged, money[['Away Team', 'TM Value', 'year']], on=['Away Team', 'year'], how='left').rename(columns={'TM Value':'tm_a'})
merged
```

```python
import numpy as np
merged['log_ratio_tm'] = np.log(merged['tm_h'] / merged['tm_a'])
merged
```
